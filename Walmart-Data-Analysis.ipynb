```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Walmart.csv")
```

```python
data
```

```python

data["Order Date"] = data["Order Date"].astype("datetime64[ns]")
data["Ship Date"] = data["Ship Date"].astype("datetime64[ns]")

```

```python
data.info()
```

```python
import datetime as dt
```

```python

data["Order Date Month"] = data["Order Date"].dt.month
data["Order Date Year"] = data["Order Date"].dt.year
```

```python
data["Order Date Year"].unique()
```

```python
data
```

```python
plt.bar(data["Quantity"],data["Profit"],color="black")
plt.title("Quantities sold VS Profit Earned",font="Lucida Console",size=18)
plt.xlabel("-----Quantity-----",font="MS Gothic",size=16)
plt.ylabel("-----Profit-----",font="MS Gothic",size=16)
```

```python
plt.bar(data["Order Date Year"],data["Profit"],color="purple")
plt.title("Change in Profit over the Years",font="Lucida Console",size=19)
plt.xlabel("-----Order Date Year-----",font="Times New Roman",size=18)
plt.ylabel("-----Profit-----",font="Times New Roman",size=18)
```

```python

plt.bar(data["Order Date Year"],data["Sales"],color="orange")
plt.title("Sales over the Year",font="Tahoma",size=25)
plt.xlabel("===== Order Date Year =====",font="Consolas",size=16)
plt.ylabel("====== Sales =====",font="Consolas",size=16)
```

```python
plt.bar(data["Category"],data["Quantity"],color="pink")
plt.title("Quantities sold according to Category",font="Lucida Console",size=20)
plt.xlabel("=====Category=====",font="Trebuchet MS",size=16)
plt.ylabel("=====Quantity=====",font="Trebuchet MS",size=16)
plt.xticks(rotation =90)
plt.show()
```

```python
plt.bar(data["Category"],data["Sales"],color="yellow")
plt.title("Sales by different categories",font="Trebuchet MS",size=20)
plt.xlabel("=====Category=====",font="MS Gothic",size=17)
plt.ylabel("=====Sales=====",font="MS Gothic",size=17)
plt.xticks(rotation=90)
plt.show()

```

```python

plt.bar(data["Category"],data["Profit"],color="cyan")
plt.title("Profit earned by different categories",font="Comic Sans MS",size=20)
plt.xlabel("=====Category=====",font="Tahoma",size=15)
plt.ylabel("=====Profit=====",font="Tahoma",size=15)
plt.xticks(rotation=90)
plt.show()
```

```python

plt.bar(data["Order Date Month"],data["Quantity"],color="magenta")
plt.title("Quantities ordered over the months",font="Franklin Gothic Medium",size=20)
plt.xlabel("=====Order Date Month=====",font="Lucida Console",size=17)
plt.ylabel("=====Quantity=====",font="Lucida Console",size=17)             
plt.show()
```

```python

plt.bar(data["Order Date Month"],data["Sales"],color="blue")
plt.title("Sales over the months",font="Franklin Gothic Medium",size=20)
plt.xlabel("=====Order Date Month=====",font="Lucida Console",size=17)
plt.ylabel("=====Sales=====",font="Lucida Console",size=17)             
plt.show()
```

```python

plt.bar(data["Order Date Month"],data["Profit"],color="green")
plt.title("Profit earned over the months",font="Franklin Gothic Medium",size=20)
plt.xlabel("=====Order Date Month=====",font="Lucida Console",size=17)
plt.ylabel("=====Profit=====",font="Lucida Console",size=17)   
plt.show()
```

```python

plt.bar(data["Order Date Year"],data["Sales"],color="yellow")
plt.bar(data["Order Date Year"],data["Profit"],color="blue")
plt.title("Sales VS Profit(Yearly)",font="Lucida Console",size=20)
plt.xlabel("=====Order Date Year=====",font="Lucida Console",size=17)
plt.ylabel("=====Sales VS Profit=====",font="Lucida Console",size=17)
plt.legend(labels=["Sales","Profit"])
plt.show()
```

```python
plt.bar(data["Order Date Month"],data["Sales"],color="yellow")
plt.bar(data["Order Date Month"],data["Profit"],color="blue")
plt.title("Sales VS Profit(Monthly)",font="Lucida Console",size=20)
plt.xlabel("=====Order Date Month=====",font="Lucida Console",size=17)
plt.ylabel("=====Sales VS Profit=====",font="Lucida Console",size=17)
plt.legend(labels=["Sales","Profit"])
plt.show()
```
